```python
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
import pandas as pd
from collections import namedtuple
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('../logs/MiniGrid-Empty-6x6-v0/dqn/1/grotile-2018-12-03-17-29-55/train/results.csv', index_col=0)
```

```python
Model = namedtuple('Model', ['name', 'train','test', 'memory'])
```

```python
model_list = ["dqn", "sql", "mirl", "gl", "qneu"]
```

```python
model = {"train":None, "test":None, "memory":None}
models={m:model.copy() for m in model_list}
```

```python
def load_trajectory(path):
    df = pd.read_pickle(path)

    keys = ("s1", "a", "r", "s2", "d")
    memory = {k:[] for k in keys}

    s1,a,s2,r,dones = [], [], [], [],[]
    for d in df:
        for idx, k in enumerate(memory.keys()):
            memory[k].append(d[idx])

    memory = {k:np.array(v) for k,v in memory.items()}
    return memory
```

```python
for path, dirs, f in os.walk('../logs'):
    dirs.sort()
    f.sort()
    try:
        name, env = path.split('/')[2:4]
    except:
        pass
    if "train" in path and 'results.csv' in f:
        print("found train")
        f = f[f.index('results.csv')]
        models[name]["train"] =  pd.read_csv(os.path.join(path,f),index_col=0)
    if "test" in path and 'results.csv' in f:
        print("found test")
        f = f[f.index('results.csv')]
        models[name]["test"] = pd.read_csv(os.path.join(path,f),index_col=0)
    if "memory.pkl" in f:
        print(f"found {name} for {env} at {path}")
        f = f[f.index('memory.pkl')]
        models[name]["memory"] = load_trajectory(os.path.join(path,f))
    else:
        pass
```

```python
models = {k:Model(name=k,**v) for k,v in models.items()}
```

```python
def smooth(x,w=0.99):
    p = x[0]
    xs = []
    for d in x:
        p = p * w + (1-w) * d
        xs.append(p)
    return np.array(xs)
```

```python
legend =[]
fig, ax = plt.subplots(1,1, figsize=(20,5))
for k, v in models.items():
    ax.plot(smooth(v.train.rw.values))
    legend.append(v.name)
ax.legend(legend)
```

```python
legend =[]
fig, ax = plt.subplots(1,1, figsize=(20,5))
for k, v in models.items():
    ax.plot(smooth(v.test.rw.values))
    legend.append(v.name)
ax.legend(legend)
```

```python
legend =[]
fig, axs = plt.subplots(len(models.keys()),1, figsize=(20,20))
for idx,(k, v) in enumerate(models.items()):
    axs[idx].plot(smooth(v.memory["a"]))
    axs[idx].set_title(v.name)
    legend.append(v.name)
ax.legend(legend)
```
